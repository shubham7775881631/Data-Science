```python
profit = [456,789,585,658,236]
```

```python
plt.subplot(2,1,1)
plt.plot(customer,sales,'red')
plt.xlabel('Customer')
plt.ylabel('Sales')
plt.title('Customer vs Sales')


plt.subplot(2,1,2)
plt.plot(customer,profit,'blue')
plt.xlabel('Cutomer')
plt.ylabel('Profit')
plt.title('Customer vs Profit')
 
plt.tight_layout()    # it is used chart to get  seprate ,clean
```

# Zoom Charts or  Assumed charts:-- 

##  1)Method :--   By Using    plt.xlim( ,   )  and    ylim( , )

```python
# By the Xlim(()) and ylim(()) funcation we can Select the data that would be vizualize from lists 

plt.plot(customer,sales,'red')
plt.xlabel('Customer')
plt.ylabel('Sales')
plt.title('Customer vs Sales')



```

```python

plt.plot(customer,sales,'red')
plt.xlabel('Customer')
plt.ylabel('Sales')
plt.title('Customer vs Sales')


# Zoom the data lies on the x axis of custmer 3 4 and 5

plt.xlim('customer3','customer5')   # it selects all the data customer 3 and 5
```

```python

plt.plot(customer,sales,'red')
plt.xlabel('Customer')
plt.ylabel('Sales')
plt.title('Customer vs Sales')

#Zoom the data Lies on y-axies

plt.ylim(5000,6000)
```

```python
plt.plot(customer,sales,'blue')
plt.xlabel('Customer')
plt.ylabel('Sales')
plt.title('Customer vs Sales')


# Zoom the data lies on the x axis of custmer 3 4 and 5

plt.xlim('customer3','customer5')   # it selects all the data customer 3 and 5


#Zoom the data Lies on y-axies

plt.ylim(5000,10000)


```

## 2) Method :-  BY Using plt.axis(( x axis  data range   ,  y axis data range ))

```python
plt.plot(customer,sales)
plt.xlabel('customers')
plt.ylabel('Sales')
plt.title('Customers vs Sales')


# here is the funcation

plt.axis(('customer3','customer5',5000,10000))
```

##  3) Method:-- by using OOPS(object Oreinted Programming  concept)

```python
# By creating an object

Graph = plt.figure()
Axis = Graph.add_axes([0.2,0.2,0.9,0.9])
Axis.plot(customer,sales)
Axis.set_xlabel('Cutomers')
Axis.set_ylabel('Sales')
Axis.set_title('Customers vs Sales')
```

```python
# Creating the two charts by creating an objects

Graph = plt.figure()
Axis1 = Graph.add_axes([0.001,0.1,0.9,0.9])
Axis1.plot(customer,sales,'red')
Axis1.set_xlabel('Cutomers')
Axis1.set_ylabel('Sales')
Axis1.set_title('Customers vs Sales')



Graph = plt.figure()
Axis2 = Graph.add_axes([1,0.1,0.9,0.9])
Axis2.plot(customer,profit,'blue')
Axis2.set_xlabel('Cutomers')
Axis2.set_ylabel('Profit')
Axis2.set_title('Customers vs Profit')
```

```python
# Creating the two charts by creating an objects in on colns

Graph,Axis = plt.subplots(nrows=2,ncols=1)

Axis[0].plot(customer,sales,'red')
Axis[0].set_xlabel('Cutomers')
Axis[0].set_ylabel('Sales')
Axis[0].set_title('Customers vs Sales')


Axis[1].plot(customer,profit,'blue')
Axis[1].set_xlabel('Cutomers')
Axis[1].set_ylabel('Profit')
Axis[1].set_title('Customers vs Profit')


plt.tight_layout()
```
