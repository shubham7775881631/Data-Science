# Creating an object by using the OOPs concept

```python
Graph, (Axis1,Axis2) = plt.subplots(nrows=2, figsize=(10,6))

sns.distplot(df['Salary'], ax = Axis1)
sns.distplot(df['Salary'], ax = Axis2)

Axis2.set(xlim=(0,10), xlabel="MySalary")
Axis2.axvline(x=0.7, label="MYtarget", linestyle='--')

Axis2.legend()
```

```python
# Chart with color


sns.distplot(df["Salary"],color='r')
```

# Styles For the Data Vizualization

```python
print(plt.style.available)
```

```python
for Style in ['white','dark','whitegrid','ticks','dakgrid']:
    sns.set_style(Style)
    sns.distplot(df['Salary'])
    plt.show()
```

```python
#creatig an rug chart more the rug more the salary

sns.distplot(df["Salary"],hist=False,rug=True)
```

```python
# to get the Shaded regeion

sns.distplot(df["Salary"],hist=False,rug=True,kde_kws={'shade':True})
```

```python
# here we can not add color

sns.regplot(data=df, x='Age',y='Salary', marker='+')
```

```python
#we can add the color

sns.lmplot(data=df, x='Age',y='Salary',hue="Department")
```

```python
# Getting the 3 differnt colors


sns.lmplot(data=df, x='Age',y='Salary',col="Department")
```

```python
# Creating the Join pot

sns.jointplot(data=df, x= 'Age', y = 'Salary')
```

```python
# Probability  Distrubustion

sns.jointplot(data=df, x='Age',y='Salary',kind='kde')
```
